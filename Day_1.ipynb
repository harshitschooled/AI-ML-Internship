```python
# Initialize encoded_data as a copy of df to avoid modifying the original DataFrame.
encoded_data = df.copy()

# Apply label encoding for 'Sex'
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
encoded_data['Sex_Label'] = label_encoder.fit_transform(encoded_data['Sex'])

# Perform one-hot encoding for 'Embarked'
one_hot_encoded = pd.get_dummies(encoded_data['Embarked'], prefix='Embarked')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
# 3. Ordinal Encoding For p class
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(encoded_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
numerical_col=['Age','Fare','Pclass']
# Corrected 'implace' to 'inplace' and applied to 'encoded_data'
encoded_data.dropna(subset=numerical_col,inplace=True)
minmax_scaler=MinMaxScaler()
# Corrected 'scaler' to 'minmax_scaler'
encoded_data[numerical_col]=minmax_scaler.fit_transform(encoded_data[numerical_col])
```

```python
# First, ensure 'Fare' is imputed in df, using its mean.
df['Fare_imputed'] = df['Fare'].fillna(df['Fare'].mean())

# Use the already imputed 'Age_mean_imputed' column.
# Create new MinMaxScaler instances for each column to scale them independently
age_scaler = MinMaxScaler()
fare_scaler = MinMaxScaler()

df['Age_MinMax'] = age_scaler.fit_transform(df[['Age_mean_imputed']])
df['Fare_MinMax'] = fare_scaler.fit_transform(df[['Fare_imputed']])
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
df['Age_Standard'] = standard_scaler.fit_transform(df[['Age_mean_imputed']])
df['Fare_Standard'] = standard_scaler.fit_transform(df[['Fare_imputed']])
```
